```python
import pandas as pd
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
```

```python
image_dataset=pd.read_excel("/content/training data.xlsx")
image_dataset.head()
```

```python
image_dataset.drop(0,inplace=True)
```

```python
features=image_dataset.drop(columns=['target','white_black_ratio'])
target=image_dataset['target']
```

```python
train_X,test_X,train_Y,test_Y=train_test_split(features,target,test_size=0.3)
```

```python
classifier=SVC(gamma='auto')
classifier.fit(train_X,train_Y)
```

```python
train_X_pred=classifier.predict(train_X)
training_data_accuracy=accuracy_score(train_Y,train_X_pred)
```

```python
print(training_data_accuracy)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic_reg=LogisticRegression()
logistic_reg.fit(train_X,train_Y)
```

```python
train_X_pred=logistic_reg.predict(train_X)
training_data_accuracy=accuracy_score(train_Y,train_X_pred)
```

```python
print(training_data_accuracy)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=200)
model.fit(train_X,train_Y)
```

```python
train_X_pred=model.predict(train_X)
training_data_accuracy=accuracy_score(train_Y,train_X_pred)
```

```python
print(training_data_accuracy)
```

```python
cv_score_lr=cross_val_score(LogisticRegression(),features,target,cv=3)
```

```python
print(cv_score_lr.mean())
```

```python
cv_score_svc=cross_val_score(SVC(kernel='linear'),features,target,cv=5)
print(cv_score_svc.mean())
```

```python
cv_score_svc=cross_val_score(SVC(kernel='sigmoid'),features,target,cv=5)
print(cv_score_svc.mean())
```

```python
cv_score_random=cross_val_score(RandomForestClassifier(),features,target,cv=50)
print(cv_score_random.mean())
```

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier()
model.fit(train_X,train_Y)
```

```python
train_X_pred=model.predict(train_X)
training_data_accuracy=accuracy_score(train_Y,train_X_pred)
training_data_accuracy=accuracy_score(train_Y,train_X_pred)
print(training_data_accuracy)
```
